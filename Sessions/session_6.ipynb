```python
import pandas as pd


url = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/1980%202023%20average%20house%20prices.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()

#need to change period to date

df['Period_dt'] = pd.to_datetime(df['Period'], format = '%Y-%m')

df.info() # now have a date column

# These columns have inconsitent capitalisation to make it easier to work with.
```

```python
# These columns have inconsitent capitalisation to make it easier to work with.

print(df.columns)

#How to overwrite these column names 
#map(access the strings and do .lower to it, do it to this selection)
df.columns = map(str.lower, df.columns)

print(df.columns) # Now they are all lowercase 
```

### Task 1 

```python
#CINreferralDate
#DateOfInitialCPC

#CPC Timeliness - difference between referral and ICPC

#cin.head()

cin['CINreferralDate_dt'] = pd.to_datetime(cin['CINreferralDate'], format = r'%d/%m/%Y', errors = 'coerce')
cin['DateOfInitialCPC_dt'] = pd.to_datetime(cin['DateOfInitialCPC'], format = r'%d/%m/%Y', errors = 'coerce')

#cin.info()


cin['CPC_Timeliness'] = cin['DateOfInitialCPC_dt'] - cin['CINreferralDate_dt']

#cin.info()
#need to get an interger 
#can get number of work days and exluding bank holidays

cin['CPC_Timeliness'] = cin['CPC_Timeliness']/pd.Timedelta(1,'d')

cin.info() # looks like it filters out those without a ICPC out as null
```

```python

timeliness_hist = px.histogram(cin, 
                               x = 'CPC_Timeliness',
                               nbins=20) #change number of bins

timeliness_hist.show()
```

```python
# Making a box plot

timeliness_box = px.box(cin,
                        x='PrimaryNeedCode', #This can be used to spilt it or can use color
                        y= 'CPC_Timeliness'
                        # ,color = 'PrimaryNeedCode'
                        )

timeliness_box.show()
```

```python
# Marginal Scatter

marginal_scatter = px.scatter( cin,
                              x = 'CINreferralDate_dt',
                              y = 'CPC_Timeliness',
                              marginal_y='violin', 
                              marginal_x='box',
                              trendline='ols')

marginal_scatter.show()

# this lets us see the most common values on each axis. Bigest y bulge is 15->16 and referrals were most common in 2020-22

#differnt types include  violin and rug and box
```

```python
## Facet Plots  How to show lots of bits of data on many graphs
#doesn;t have to be histograms

facet_plot = px.histogram(cin,
                          x='ReferralSource',
                          facet_col='PrimaryNeedCode' #,
                          #facet_row='ReasonForClosure'
                          )

facet_plot.show()

# its like small multiples 

# much better examples https://plotly.com/python/facet-plots/
```

```python
cin_no_nans = cin[cin['CPC_Timeliness'].notna()]

size_scatter = px.scatter(cin_no_nans,
                          x='CINreferralDate_dt',
                          y='PrimaryNeedCode',
                          size='CPC_Timeliness')

size_scatter.show()
```
