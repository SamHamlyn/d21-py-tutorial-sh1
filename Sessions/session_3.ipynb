## Task - Transforming some data 
We want to add age to the c_identifiers table as of 31/03/2026

```python
c_identifiers

# we will use PersonBirthDate_dt
#https://www.geeksforgeeks.org/python/get-month-and-year-from-date-in-pandas-python/

c_identifiers["PersonBirthDate_dt"]

collectiondate = pd.to_datetime("31/03/2026", format = "%d/%m/%Y")

c_identifiers["age_delta"] = collectiondate - c_identifiers["PersonBirthDate_dt"]

c_identifiers["age_delta"] # we have a list of timedetlas 

c_identifiers["yrs"] = pd.DatetimeIndex(c_identifiers["PersonBirthDate_dt"]).year
c_identifiers["dayofyr"] = pd.DatetimeIndex(c_identifiers["PersonBirthDate_dt"]).dayofyear

collection_yr = 2026
collection_dayofyr = 90 #from wolfram alpha 

c_identifiers['Age'] = collection_yr - c_identifiers["yrs"]  -(collection_dayofyr < c_identifiers["dayofyr"])
```
