```python
# Now I do it with the gravity database 

filepath = "/workspaces/d21-py-tutorial-sh1/Sessions/Second Series/data/gravity.db"
enginedrivers = "sqlite+pysqlite:///"
engine_gravity = create_engine(enginedrivers+filepath)

connection_gravity = engine_gravity.connect() 

inspection_gravity = inspect(connection_gravity) 

inspection_gravity.get_table_names()
```

```python
#a meta data object stores the information about the data. 

metadata_903 = MetaData() # this makes the object

metadata_903.tables.keys() # currently an empty dictionary

episodes_903 = Table('episodes', metadata_903, autoload_with = engine_903)

print(episodes_903.name)
print(episodes_903.c.keys()) # just doing .c would have tablename.column name
print(repr(episodes_903)) # repr() gives more informaiton like data dypes
```

```python
# now do it with the book table 

metadata_gravity = MetaData()
#metadata_gravity.tables.keys()
book_gravity = Table('book', metadata_gravity, autoload_with = engine_gravity)

print(book_gravity.name)
print(book_gravity.c.keys()) # just doing .c would have tablename.column name
print(repr(book_gravity))
```

```python
#with sql alchmeny you can query stuff with sql queries. but you can also do it with python style queires. it will translate it for you

#we don't want to keep engines open for long time as it would use lots of memort. therefore we use a with block
#variable names will be kept only in the with block. makes it easier to write

with engine_903.connect() as con: # only in with witll enginge.connect be called con
    stmt = "SELECT * FROM episodes"
    result_proxy = con.execute(text(stmt)) # this is an intermediate wher we we chose which bits of the data we want to use and tell it how query written (sql text)
    result = result_proxy.fetchall() #give us all the data from the query

first_row = result[0]
print(first_row)
print(first_row.CHILD)
```

```python
# do it with books 

with engine_gravity.connect() as con:
    stmt = "select * from book"
    result_proxy = con.execute(text(stmt))
    result = result_proxy.fetchmany( size = 10) # can use fetch many to detirmine how many rows to get

print(result)
first_row_gravity = result[0]
print(first_row_gravity)
print(first_row_gravity.title)
```

```python
#'can do query in python style query
with engine_903.connect() as con:
    stmt = select(episodes_903)
    #print(stmt)
    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=5)
print(result)
```

```python
with engine_gravity.connect() as con:
    stmt = select(book_gravity)
    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=5)
print(result)
```

```python
header = Table("Header", metadata_903, autoload_with=engine_903) # make a new table

with engine_903.connect() as con:
    stmt = select(header).where(or_(header.columns.ETHNIC == "REFU",
                                    header.columns.ETHNIC == "NOBI"))  #stmt = statement # adding condtions in a python/pandas way. they use or_ for or as or already used
    #print(stmt)
    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=16)

print(result)
```

```python
#all from books when publication_date is eiter 1999 or 2000
import datetime

with engine_gravity.connect() as con:
    stmt = select(book_gravity).where(and_(book_gravity.columns.publication_date >= datetime.datetime(1999,1,1),
                                    book_gravity.columns.publication_date <= datetime.datetime(2000,1,1)))  #stmt = statement # adding condtions in a python/pandas way. they use or_ for or as or already used
    #print(stmt)
    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=16)

print(result)
```
